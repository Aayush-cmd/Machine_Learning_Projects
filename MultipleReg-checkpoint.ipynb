```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
x = df[['hp','wt','gear']] #input
y = df['mpg']        #output
```

```python
x
```

```python
y
```

# feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
pd.DataFrame(x_scaled).describe()
```

## building model



```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_scaled,y)
```

```python
lr.score(x_scaled,y)
```

```python
lr.coef_
```

```python
lr.intercept_                 #constant from eqn of line
```

```python
hp  =130
wt  = 4.1
gear = 5
lr.predict(scaler.transform([[hp,wt,gear]]))
```
