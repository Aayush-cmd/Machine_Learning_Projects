```python
import pandas as pd 
```

```python
data=[['apple','bear','rice','chicken'],
      ['apple','bear','rice'],
      ['apple','bear'],
      ['apple','pear'],
      ['apple','bear','rice','chicken'],
      ['apple','bear','rice'],
      ['apple','bear'],
      ['apple','pear']]
```

### Transaction encoder for creating structured data

```python
!pip install mlxtend
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
```

```python
df_enc=te.fit_transform(data)
```

```python
df_enc
```

```python
te.columns_
```

```python
df=pd.DataFrame(df_enc,columns= te.columns_)
```

```python
df
```

## Generate frequent itemset


```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_itemset =apriori(df,min_support=0.25,use_colnames=True)
```

```python
freq_itemset
```

### generate the association rules

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules=association_rules(freq_itemset,metric='confidence',min_threshold=0.50)
```

```python
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence']]
```

```python
rules
```

### Extract using condition

```python
rules[(rules['confidence']>0.5)&(rules['support']>0.25)]
```

### Recomendation

```python
rules[rules['antecedents']=={'rice'}]
```

### Sort

```python
rules.sort_values(by='confidence',ascending=False);
```

```python
# Save the Rules

rules.to_csv('rules_csv',index=False)
```
