```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
x = df.iloc[:,[3,4]]
```

```python
x
```

# Dendrogram

```python
from scipy.cluster.hierarchy import dendrogram , linkage
```

```python
plt.figure(figsize = (16,9))
dendrogram(linkage(x,'complete'));
```

```python
#form clusters
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag= AgglomerativeClustering(n_clusters = 5)
```

```python
labels = ag.fit_predict(x)
```

```python
labels
```

```python
plt.figure(figsize= (16,9))
plt.subplot(1,2,1)
plt.style.use('dark_background')
plt.title('Unclustered Data')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score ')
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],marker='o')
plt.grid()

plt.subplot(1,2,2)
plt.style.use('dark_background')
plt.title('Unclustered Data')
plt.xlabel('Annual Income ')
plt.ylabel('Spending Score ')
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],marker='o',c = labels)
plt.grid()
```
