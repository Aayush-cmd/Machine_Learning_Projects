```python
from imblearn.over_sampling import SMOTE
```

```python
smo =SMOTE(random_state =0)
```

```python
x_res,y_res = smo.fit_resample(x_new,y_new)
```

```python
x_res.shape
```

```python
y_res.shape
```

```python
y_res.value_counts()
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
skb = SelectKBest(k=5,score_func = chi2)
```

```python
skb.fit(x_res,y_res)
```

```python
skb.get_support()
```

```python
x = x_res.iloc[:,skb.get_support()]
```

```python
x.describe()
```
