```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('BostonHousing.csv')
```

```python
df.head()
```

```python
x=df.drop('medv',axis = 1) #Input
y=df['medv']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
sca= StandardScaler()
```

```python
x_scaled =sca.fit_transform(x)
```

### cross validation

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
x_train.shape
```

### Build linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
lr.score(x_train,y_train)
```

### Poly Regg

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
x_poly=poly.fit_transform(x_scaled)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_poly,y,random_state=0,test_size=0.2)
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print('Training acc',lr.score(x_train,y_train))
print('Testing acc',r2_score(y_test,y_pred))
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=15)
```

```python
ridge.fit(x_train,y_train)
```

```python
print('Training acc',ridge.score(x_train,y_train))
print('Testing acc',ridge.score(x_test,y_test))
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=20)
ridge.fit(x_train,y_train)
```

```python
print('Training acc',ridge.score(x_train,y_train))
print('Testing acc',ridge.score(x_test,y_test))
```

### LassoReg

```python
from sklearn.linear_model import Lasso
las=Lasso(alpha=0.2)
las.fit(x_train,y_train)
```

```python
print('Training acc',las.score(x_train,y_train))
print('Testing acc',las.score(x_test,y_test))
```
